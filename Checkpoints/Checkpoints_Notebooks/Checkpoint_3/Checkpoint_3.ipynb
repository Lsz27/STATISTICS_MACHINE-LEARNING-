```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
df_vendas = pd.read_csv('online_retail_small.csv', sep = ';', encoding ='latin1', decimal = ',')
df_vendas
```

```python
df_vendas.columns = ['n_pedido', 'stockCode', 'descrição', 'quantidade', 'data', 'preço unitario', 'id_cliente', 'pais']
df_vendas
```

###1 –Quantos clientes diferentes existem no banco de dados?

```python
uniqueId = len(df_vendas["id_cliente"].unique())
uniqueId
```

##2 –Quantos clientes foram observados em cada pais?

```python
# Contar quantos clientes foram observados em cada país
clientes_por_pais = df_vendas['pais'].value_counts()

# Exibir o resultado
print(clientes_por_pais)
```

##3 Quantas compras diferentes cada cliente realizou? Quais os top 10?






```python
df_vendas
```

```python
# Contar quantas compras diferentes cada cliente realizou
compras_por_cliente = df_vendas.groupby('id_cliente')['n_pedido'].nunique()

compras_por_cliente
```

```python
top_10_clientes = compras_por_cliente.nlargest(10)

print(top_10_clientes)
```

##4 –Quais  os  top  10 clientes  que  mais  compraram  em  quantidades? Quais  os  top  10  clientes que mais compraram em valor?

```python
df_vendas
```

```python
# Top 10 clientes que mais compraram em quantidades
top_10_quantidade = df_vendas.groupby('id_cliente')['quantidade'].sum().nlargest(10)
top_10_quantidade
```

```python

# Somar os valores da coluna 'Valor'
valor_total = df_vendas['preço unitario'].sum()

print(f"A soma dos valores é: {valor_total}")
```

```python
# Top 10 clientes que mais compraram em valor
df_vendas['valor_total'] = df_vendas['quantidade'] * df_vendas['preço unitario']  

# Criar coluna com valor total por compra
top_10_valor = df_vendas.groupby('id_cliente')['valor_total'].sum().nlargest(10)
```

```python
top_10_valor
```

##5 –Um dos produtos possui a seguinte função lucro −𝟐𝒑𝟐+𝟓𝟎𝒑−𝟏𝟎. Encontre a derivada da função lucro. Qual seria o melhor preço?

```python
import pandas as pd
from sympy import symbols, diff



# Função de lucro
p = symbols('p')
funcao_lucro = -2 * p**2 + 50 * p - 10

# Derivada da função de lucro
derivada_lucro = diff(funcao_lucro, p)

# Encontrar o valor de p para o qual a derivada é igual a zero
melhor_preco = max(df_vendas['preço unitario'])

# Exibir os resultados
print("Derivada da função de lucro:")
print(derivada_lucro)
print("\nMelhor preço:")
print(melhor_preco)
```
