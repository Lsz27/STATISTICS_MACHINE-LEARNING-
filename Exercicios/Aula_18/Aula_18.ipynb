```python
import numpy as np
import pandas as pd
from scipy import stats as st
```

```python
dados_salarios = pd.read_csv('salarios_profs.csv', encoding= 'latin1', sep = ';', decimal = ',')
```

```python
dados_salarios.sort_values('SALARIO_MENSAL', ascending= False)
```

```python
dados_profs = dados_salarios[(dados_salarios['FUNCAO'] == 'Professor Titular')&\
               (dados_salarios['JORNADA']== 'RDIDP')&\
               (dados_salarios['UNID_ORGAO']== 'ICMC')]
```

```python
dados_profs.groupby('UNID_ORGAO')['SALARIO_MENSAL_LIQUIDO'].describe()
```

```python
st.ttest_1samp(a = dados_profs['SALARIO_MENSAL_LIQUIDO'],
               popmean = 2000,
               alternative = 'greater')
```

```python
dados_profs = dados_salarios[(dados_salarios['FUNCAO'] == 'Professor Titular')&\
               (dados_salarios['JORNADA']== 'RDIDP')&\
               (dados_salarios['UNID_ORGAO'].isin(['ICMC', 'FEA']))]
```

```python
dados_profs.groupby('UNID_ORGAO')['SALARIO_MENSAL_LIQUIDO'].describe()
```

```python
gra = dados_profs[dados_profs['UNID_ORGAO']== 'ICMC']['SALARIO_MENSAL_LIQUIDO']
grb = dados_profs[dados_profs['UNID_ORGAO']== 'FEA']['SALARIO_MENSAL_LIQUIDO']
```

```python
st.ttest_ind(a = gra, b = grb, alternative = 'two-sided')
```
