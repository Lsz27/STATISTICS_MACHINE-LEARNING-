#Exemplo 2

* Regra da potência 
- $\frac{dx^2}{dx}$

```python
x = sp.Symbol("x")
x
```

```python
funcao_original = x**2
funcao_original
```

```python
funcao_derivada = sp.diff(funcao_original)
funcao_derivada
```

$\frac{3x^3 + 2x^2 + x}{dx}$

```python
x = sp.Symbol("x")
funcao_original = 3*x**3 + 2*x**2 + x
funcao_original
```

```python
# Não aceita numeros reais
funcao_derivada = sp.diff(funcao_original)
funcao_derivada
```

```python
sp.plot(funcao_original)
```

```python
sp.plot(funcao_derivada)
```

```python
# erro, objetivo simbolico, conseguimos utilizar a função lambdfy 
#funcao_original(2) 
```

* Lambdify

```python
funcao_original_calc = sp.lambdify(x, funcao_original)
```

```python
funcao_original_calc(2)
```

```python
funcao_derivada_calc = sp.lambdify(x, funcao_derivada)
```

```python
funcao_derivada_calc(2)
```

### Exemplo 3 
- $\frac{d(x^2+3)\cdot(2x1)}{dx}$

```python
x = sp.Symbol("x")
funcao_original = (x**2+3)*(2*x+1)
funcao_original
```

```python
funcao_derivada = sp.diff(funcao_original)
funcao_derivada
```

```python
funcao_derivada = sp.simplify(funcao_derivada)
funcao_derivada
```

```python
#funcao_derivada(2)
```

```python
funcao_derivada_calc = sp.lambdify(x, funcao_derivada)
```

```python
funcao_derivada_calc(2)
```

#Equação de lucro

$Lucro = -2p^2 = 50,4p -10$

```python
p = sp.Symbol("p")
```

```python
funcao_original = -2*p**2 + 50.4*p -10
funcao_original
```

```python
funcao_original_calc = sp.lambdify(p, funcao_original) 
```

```python
funcao_original_calc(15)
```

```python
x = [i for i in range(20)]
y = [funcao_original_calc(i)for i in range(20)]
plt.plot(x ,y)
```
