## Exemplos Permutação e Combinação

### Permutação

- Sem repetição

```python
alunos = ['biel', 'tiago', 'cyrus']
```

```python
posicoes_bolsa = permutations(alunos, 3)
```

```python
# permutacoes - estrategia 1
lista_permutacoes = list(posicoes_bolsa)
lista_permutacoes
```

```python
# permutacoes - estrategia 2
posicoes_bolsa = permutations(alunos, 3)
for per in posicoes_bolsa:
  print(per)
```

- Com repetição

```python
posicoes_bolsa = product(alunos, repeat = 3)
list(posicoes_bolsa)
```

- Quantidade de permutações

```python
# quantas permutacoes - estrategia 1
len(lista_permutacoes)
```

```python
# estrategia 2
def formula_permutacao(n, k, repeat = True):
  if repeat is True:
    # com repeticao
    return n**k
  else:
    # sem repeticao
    return math.factorial(n) / math.factorial(n-k)
```

```python
# com repeticao
formula_permutacao(3,3)
```

```python
# sem repeticao
formula_permutacao(3,3, repeat = False)
```

```python

# estrategia 3 - so funciona para sem repetição
# detalhes: https://docs.python.org/3/library/math.html
math.perm(3,3)
```

## Combinação

- Sem repetição

```python
alunos = ['biel', 'tiago', 'cyrus', 'isabela', 'larissa']
alunos
```

```python
grupos = combinations(alunos, 2)
list(grupos)
```

- Com repetição

```python
grupos = combinations_with_replacement(alunos, 2)
list(grupos)
```

- Quantidade de combinações

```python
# estrategia 1
grupos = combinations(alunos, 2)
len(list(grupos))
```

```python
# estrategia 2
def formula_combinacao(n, k, repeat = True):
  if repeat is True:
    # com repeticao
    return math.factorial(n+k-1) / (math.factorial(k) * math.factorial(n-1))
  else:
    # sem repeticao
    return math.factorial(n) / (math.factorial(k) * math.factorial(n-k))
```

```python
# com repeticao
formula_combinacao(5, 2)
```

```python
# sem repeticao
formula_combinacao(5, 2, repeat = False)
```

```python
# estrategia 3 - funciona para sem repeticao
math.comb(5,2)
```
